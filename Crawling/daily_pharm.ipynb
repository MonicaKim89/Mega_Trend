```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import NoSuchElementException
import selenium.common.exceptions as sex

import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
import lxml

import time
from time import sleep
import random
import re
```

```python
search_url = []

for i in range(1,344):
    urls = ('http://www.dailypharm.com/Users/News/NewsSearch.html?nPage={}&dpsearch=%B9%E9%BD%C5&mode=news&nSection=&nSubSection=&stDate=2018-01-01&edDate=2022-04-14'.format(i))
    search_url.append(urls)
```

```python
path = "C:\\chromedriver.exe"
driver = webdriver.Chrome(path)
driver.implicitly_wait(5)

start_url = "http://www.dailypharm.com/Users/News/NewsSearch.html?nPage=11&dpsearch=%B9%E9%BD%C5&mode=news&nSection=&nSubSection=&stDate=2018-01-01&edDate=2022-04-14"
driver.get(start_url)
```

```python
title_list = []
dates_list = []
url_list =[]

z = 0

for i in search_url:
    driver.get(i)
    page = driver.page_source
    soup = BeautifulSoup(page, 'html.parser')

    title = soup.find_all('div', class_='listHead')
    for i in title:
        k = i.text
        title_list.append(k)
        dates_list.append(k[-19:])


    urls = soup.find_all('li',attrs = {'class':'newsList'})

    for i in urls:
        address = i.a['href']
        address = 'http://www.dailypharm.com/Users/News/'+address
        url_list.append(address)

    z +=1

    print('----------------------------', len(search_url)-z , '남음-------------------------')
```

```python
print(len(title_list))
print(len(dates_list))
print(len(url_list))
```

```python
daily_pharm_v1= pd.DataFrame({'title':title_list, 'dates':dates_list, 'url':url_list})
daily_pharm_v1.head()
```

```python
daily_pharm_v1.to_excel('daily_pharm_v1.xlsx')
```

```python
text_list=[]
sub_text_list=[]

z = 0

for i in daily_pharm_v1['url']:
    driver.get(i)
    page = driver.page_source
    soup = BeautifulSoup(page, 'html.parser')


    text = soup.find_all('div', class_='newsContents font1')
    for i in text:
        k = i.text
        k = k.strip()
        text_list.append(k)

    sub_text = soup.find_all('div', class_ ='subTitleNewsBody')
    for i in sub_text:
        sub_text_list.append(i.text.strip())

    z+=1

    random_num = random.randint(1,3)
    time.sleep(random_num)

    print('----------------------------', len(daily_pharm_v1['url'])-z , '남음-------------------------')

```

```python
daily_pharm_v1['subtext'] = sub_text_list
daily_pharm_v1['text'] = text_list
```
